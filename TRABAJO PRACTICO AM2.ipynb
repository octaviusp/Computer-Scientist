```python
# IMPORTANDO LIBRERIAS NECESARIAS
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

# CURVA DE NIVEL


image.png

Como la función en el primer término es polinómica, es continua siempre.
Sin embargo, en el término dos, tiene una raíz cuadrada.
Por lo tanto, 

$\mathrm{4}-{x}^{2} > 0$

$\mathrm{x}^{2} < 4$

$\sqrt{{{x}^{2}}} < \sqrt{4}$

$\mathrm x < 2$

Entonces, nuestra función estará definida siempre y cuando X < 2.

```python
# CURVAS DE NIVEL
# creo mi funcion
my_function = lambda x,y: (x**2)*y + (4 -x**2)**(1/2)
 
# recinto de x
x = np.linspace(-2, 2, 10)
# recinto de y
y = np.linspace(-2, 2, 10)
# todos los pares ordenados posibles entre X e Y
X,Y = np.meshgrid(x, y)
# evaluo todos los Z posibles para cada X,Y
F = my_function(X,Y)

def formatter(x):
  s = f"{x:.1f}"
  if s[-1] == "0":
      s = f"{x:.0f}"
  return rf"{s}" if plt.rcParams["text.usetex"] else f"{s}"
# muestro el contorno por cada valor de la funcion, osea la curva de nivel
fig, ax = plt.subplots()
CS = ax.contour(X, Y, F)
# mostrar grafico
ax.clabel(CS, CS.levels, inline=True, fmt=formatter, fontsize=10)
plt.show()
```

# SUPERFICIE

```python
# creo mi funcion
my_function = lambda x,y: (x**2)*y + (4 -x**2)**(1/2)
 
# recinto de x
x = np.linspace(-2, 2, 50)
# recinto de y
y = np.linspace(-2, 2, 50)
# todos los pares ordenados de los recintos X e Y
X,Y = np.meshgrid(x, y)
# evaluo todos los valores de Z por cada par ordenado de X e Y
F = my_function(X,Y)

# Creo una figura con el tamaño 12,8
FIGURE = plt.figure(figsize = [12,8]) # creando la figura con su tamaño inicial
ax = FIGURE.gca(projection = "3d") # poniendo la proyeccion en 3d
ax.plot_surface(X,Y,F) # muestro la superficie con los valores correspondientes
# de x, y, f
plt.show() # mostrar grafico
```
